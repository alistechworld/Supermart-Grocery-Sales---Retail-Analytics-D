# Time Series Analysis of Sales Over Time 

```python
df.head(1)
```

```python
category_sales = df.groupby("CATEGORY")["SALES"].sum()
category_sales.plot(kind="bar", color = "skyblue")
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

# Distribution of Profit

```python
plt.hist(df["PROFIT"], bins = 10, color = "skyblue", edgecolor = "green")
plt.title("Distribution of Profit")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()
```

# Sales by Product Category

```python
category_sales = df.groupby("CATEGORY")["SALES"].sum().sort_values()
category_sales.plot(kind="bar", color = "teal")
plt.title("Total sales by product category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

# Profit vs. Discount Relationship

```python
# alpha = 0.5 used for the transparanc 
plt.scatter(df['DISCOUNT'], df["PROFIT"], alpha=0.10, color = "purple")
plt.title("Discount vs. Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
```

# Sales and Profit Distribution by Region

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='REGION', y='SALES', data=df, palette='Set3')
plt.title('Sales Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='REGION', y='PROFIT', data=df, palette='Set2')
plt.title('Profit Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.show()
```
